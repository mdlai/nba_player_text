```python
import requests
import string
import csv
import re

from bs4 import BeautifulSoup as BS
```

```python
BASEURL='http://www.draftexpress.com'
```

```python
def url_to_soup(url):
    return BS(requests.get(url).text, 'lxml')

def profile_url(soup, row):
    return soup.find_all('tbody')[0].find_all('tr')[row].find_all('td')[4].find('a')['href'].replace('\'','')

def profile_data(prof_url):
    '''
    Gets the data from the a specific profile url
    Returns array with [Standing Reach, Wingspan, Max Vertical, Text]
    '''
    profile = url_to_soup(BASEURL+prof_url)
    prof_data = []
    try:
        [s.extract() for s in profile('i')]
    except:
        pass
    try:
        regex = re.compile('\nCurrent: ([a-zA-Z]*/*[a-zA-Z]*)')
        prof_data.append(regex.match(profile.find_all(attrs={"data-title":"POSITIONS"})[0].text).group(1))
    except:
        prof_data.append('')
    try:
        prof_data.append(profile.find_all(attrs={"data-title":"Standing"})[0].text.encode('utf-8'))
    except:
        prof_data.append('')
    try:
        prof_data.append(profile.find_all(attrs={"data-title":"Wingspan"})[0].text.encode('utf-8'))
    except:
        prof_data.append('')
    try:
        prof_data.append(profile.find_all(attrs={"data-title":"Max"})[0].text.encode('utf-8'))
    except:
        prof_data.append('')
    
    output = []
    for i in range(len(profile.find_all('p'))):
        output.append(profile.find_all('p')[i].text.replace('\n',"").replace('\r',"").encode('utf-8'))
    output = ' '.join(output)
    prof_data.append(output)
    
    return prof_data

def base_data(soup, row):
    '''
    Gets the data for a single row(player) including the profile page for that player.
    Returns an array
    '''
    data = []
    entry = soup.find_all('tbody')[0].find_all('tr')[row].find_all('td')
    for i in range(len(entry)):
        data.append(entry[i].text.encode('utf-8'))
    return data + profile_data(profile_url(soup, row))

def player_to_csv(player):
    with open('players.csv','ab') as csvfile:
        players = csv.writer(csvfile)
        players.writerow(player)

def all_players(soup):
    for i in range(1,len(soup.find_all('tbody')[0].find_all('tr'))):
        player = base_data(soup,i)
        print 'Player: {}, index: {}'.format(player[4],i)
        player_to_csv(player)
```

```python
url = 'http://www.draftexpress.com/nba-draft-history/?syear=2000&eyear=2015&spick=1&epick=239&team=0&sort=0&complex=1&submit=Submit'
souped = url_to_soup(url)
```

```python
with open('players.csv','wb') as csvfile:
    movies = csv.writer(csvfile)
    movies.writerow([u'Year',
                     u'Pick',
                     u'Drafted',
                     u'Traded to',
                     u'Name',
                     u'Height',
                     u'Weight',
                     u'College',
                     u'Birthday',
                     u'Comment',
                     u'RSCI',
                     u'Position',
                     u'StandingReach',
                     u'Wingspan',
                     u'Vertical',
                     u'Text'])
```
